```python
import re
import pandas as pd
```

```python
lines = [line.strip() for line in open('xom-all-flat-mod-pnums.xml', 'r').readlines()]
lines = re.split(r'<', ' '.join(lines))
tags = []
results = []
for line in lines[1:]:
    (tag, txt) = re.split('>', line)

    # Skip pragmas
    if re.search(r'^\s*[?!]', tag):
        continue
    
    # Handle end and start tags
    if re.search(r'^\/', tag):
        tags = tags[:-1]
    else:
        tagparts = re.split(r'\s+', tag)
        tags.append(tagparts[0])
    
    # Handle singletons
    if re.search(r'.+\/$', tag):
        results.append(('/'.join(tags).strip('/'), ''))
        tags = tags[:-1]

    # Print xpath
    results.append(('/'.join(tags), txt))
```

```python
df = pd.DataFrame(results, columns=['path', 'text'])
```
